```python
import pandas as pd
import numpy as  np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline

data = pd.read_csv('titanic_train.csv')

data.head()
```

```python
#とりあえずデータの処理をする
#欠損値の確認
data.isnull().sum()
```

```python
#Embarkedの欠損部分は少数なので排除
data.dropna(subset=['Embarked'])
#Embarkedをダミー変数で代替
edummies=pd.get_dummies(data['Embarked'])
data=data.drop('Embarked', axis=1)
data=pd.concat([data,edummies],axis=1)
data.head(20)
#性別は2通りしかなくマッピングで差し支えないので、0と1で代替
sex_mapping = {'female':0,'male':1}
data['Sex'] = data['Sex'].map(sex_mapping)
#Cabinは欠損値の方が多いので説明変数そのものを削除する
data=data.drop('Cabin',axis=1)
#ID、名前、チケット番号は生死との関係が薄いと考えて削除する
data=data.drop(['PassengerId','Name','Ticket'],axis=1)
data.head(20)
```

```python
#年齢は相関のある他の変数がないので平均値補完を行う
data=data.fillna(data.Age.mean())
data.head(20)
```

```python
#データを学習用にnparrayに変換
y=data['Survived'].values
X=data.drop('Survived',axis=1).values
#データの標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_std = scaler.transform(X)
#trainデータとtestデータに分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0)
#ランダムフォレストでやってみる
from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(random_state=0, n_estimators=10)
rfc_1.fit(X_train, y_train)
```

```python
#正答率の確認
print('train accuracy: %.3f' % rfc_1.score(X_train, y_train))
print('test  accuracy: %.3f' % rfc_1.score(X_test, y_test))
```
